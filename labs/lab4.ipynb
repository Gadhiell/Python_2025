### Leer y mostrar datos de pokemon

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pkmn = pd.read_csv("All_Pokemon.csv", index_col="Number")
```

```python
pkmn.head()
```

### Identificacion de duplicados

```python
pkmn.info()
```

```python
dupla = pkmn.duplicated().sum()
if dupla > 0:
    print(f"Se encontraron {dupla} filas duplicadas, vamos a eliminarlos")
    pkmn.drop_duplicates(inplace=True)
else:
    print("No se encontraron duplicados, no es necesario eliminarlas")
```

### Identificar y contar nulos

```python
nulos_columnas = pkmn.isnull().sum() #isnull solo identifica los nulos
nulos_filas = pkmn[pkmn.isnull().any(axis=1)]
total_nulos =  nulos_columnas.sum()
print (f"esta es la cantidad total de nulos en el dataset: {total_nulos}")
nulos_filas
```

```python
pkmn_SinNulo = pkmn.dropna()
pkmn_SinNulo
```

### Filtrar Pokemon por tipo fuego

```python
def filtro_fuego(Type_1):
    return "Fire" in Type_1

#Se aplica el filtro utilizando apply
Fuego = pkmn_SinNulo[pkmn_SinNulo['Type 1'].apply(filtro_fuego)]
Fuego.head(20)
```

### Promedio Nivel de defensa (fuego)

```python
PKMn = pkmn_SinNulo[pkmn_SinNulo["Type 1"].apply(lambda x: 'Fire' in x)]["Def"]
promedio_Fire = PKMn.mean()

promedio_Fire
```
